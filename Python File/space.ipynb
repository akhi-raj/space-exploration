```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Group by country and calculate average success rate
country_success = df.groupby('Country')['Success Rate (%)'].mean().sort_values(ascending=False)

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(x=country_success.values, y=country_success.index, palette='viridis')
plt.title('Average Mission Success Rate by Country')
plt.xlabel('Average Success Rate (%)')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```
